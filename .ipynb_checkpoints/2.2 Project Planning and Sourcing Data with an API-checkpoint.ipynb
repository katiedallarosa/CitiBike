### Obtain weather data from New York LaGuardia’s weather station for 2022 and export it to a CSV file.

```python
# Define NOAA token

Token = 'fHTnfzvmHHVDRAkzJwUmixvMXCduyMfA'
```

```python
# Get the API 

r = requests.get('https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND&datatypeid=TAVG&limit=1000&stationid=GHCND:USW00094846&startdate=2022-01-01&enddate=2022-12-31', headers={'token':Token})
```

```python
# Load the api response as a json

d = json.loads(r.text)
```

```python
d
```

```python
# Secure all items in the response that correspond to TAVG

avg_temps = [item for item in d['results'] if item['datatype']=='TAVG']
```

```python
# Get only the date field from all average temperature readings

dates_temp = [item['date'] for item in avg_temps]
```

```python
# Get the temperature from all average temperature readings

temps = [item['value'] for item in avg_temps]
```

```python
temps
```

```python
# Put the results in a dataframe

df_temp = pd.DataFrame()
```

```python
# Export results to csv file

df_temp.to_csv('nytemps_data.csv')
```

### Merge the weather data with the New York CitiBike data set and export it to a CSV file.

```python
# Get only date and cast it to date time; convert temperature from tenths of Celsius to normal Celsius

df_temp['date'] = [datetime.strptime(d, "%Y-%m-%dT%H:%M:%S") for d in dates_temp]
df_temp['avgTemp'] = [float(v)/10.0 for v in temps]
```
