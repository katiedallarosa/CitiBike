```python
import pandas as pd
from keplergl import KeplerGl
df_plot = pd.read_csv('/content/drive/MyDrive/df_plot.csv')
```

```python
# Create KeplerGl instance
m = KeplerGl(height = 700, data={"data_1": df_plot})
```

```python
from google.colab import output
output.enable_custom_widget_manager()
```

```python
m
```

Using the large dataframe, we created a smaller group based on the top 100 (most common) trips for CitiBike. I used the color red for the start and end stations with a gradient red for the trip arcs based on the number of trips. The layer was not necessary due to already creating the top 100 trips, but I added it just in case.

```python
# Create config object and save the output

config = m.config
```

```python
# Export the map

m.save_to_html(file_name='CitiBike Trips Aggregated.html', read_only = False, config = config)
```

```python
# Embed config file as JSON

import json
with open("config.json", "w") as outfile:
    json.dump(config, outfile)
```
